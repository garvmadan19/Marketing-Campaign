- <b>The income dtype is non-numeric. We need to convert it into float by removing comma and dollar sign.</b>

```python
df['Income']= df['Income'].str.replace('$','')
df['Income']= df['Income'].str.replace(',','')
df['Income']= df['Income'].astype(float)
```

# Checking and filling nulls

```python
df.isnull().sum()
```

- <b>There are 24 nulls in income column. We need to treat them with their similar education background and marital status. For that we need to find the average incomes on various education and marital status.</b>

```python
df_income = df.groupby(['Education','Marital_Status'])["Income"].mean()
```

```python
df_income 
```

```python
df_income = pd.DataFrame(df_income).reset_index()
```

```python
df_income
```

- <b>After finding the average incomes we have to merge the two dataframe and have to replace null by second income column.</b>

```python
df2 = pd.merge(df,df_income, on=['Education','Marital_Status'] , how='left')
```

```python
df2
```

```python
df2['Income_x'] = np.where(df2['Income_x'].isnull(),df2['Income_y'], df2['Income_x'])
```

```python
df2.isnull().sum()
```

- <b>All the 24 nulls have been treated with average incomes based on education and marital status.</b>

- <b>Since the original df has only one income column, we have to put the new dataframe back in original dataframe by dropping the second income column and renaming first income column.</b>

```python
df = df2.drop('Income_y',axis=1)
df = df.rename(columns ={'Income_x' : 'Income'})
```

- <b>Now lets check whether the data has been correctly transferred or not.</b>

# Hypothesis Testing

## Older vs Younger Generation

- <b>For testing generationwise, first we need to put bins in age and put them in groups.</b>

```python
bins = [0,18,32,60,140]
```

```python
df['age_bucket'] = pd.cut(df['Age'],bins)
```

```python
df[['Age','age_bucket']]
```

```python
df['age_group'] = pd.cut(df['Age'],bins)
df['Web_Purchase'] = df['NumWebPurchases']*100/df['Total_Purchases']
df['Store_Purchase'] = df['NumStorePurchases']*100/df['Total_Purchases']
df['Catalogue_Purchase'] = df['NumCatalogPurchases']*100/df['Total_Purchases']
```

```python
df.groupby("age_group")[["Web_Purchase","Store_Purchase", "Catalogue_Purchase"]].mean()
```

```python
df.groupby('age_bucket')[['Web_Purchase','Store_Purchase', 'Catalogue_Purchase']].mean().round(2)
```

- <b>From the data there is no concrete evidence that older people are lesser tech savvy and are purchaing more from the store.</b>

## Kids vs No Kids

```python
df.groupby('Kidhome')[['Web_Purchase','Store_Purchase', 'Catalogue_Purchase']].mean().round(2) 
```

- <b>It is clearly visible that the people with kids are making more web purchases. So, the hypothesis made seems to be true.</b>

## Cannibalization

```python
df.groupby("age_group")[["Web_Purchase","Store_Purchase", "Catalogue_Purchase"]].mean()
```

```python
df.groupby('Kidhome')[['Web_Purchase','Store_Purchase', 'Catalogue_Purchase']].mean()
```

- <b>From the above data, we can conclude that cannibalization is happening since 50% or more than 50% sales are happening from non-store channels.</b>

## USA vs rest of the world

```python
df.groupby('Country_US')['Total_Purchases'].mean()
```

- <b>The purchase of USA is more than rest of the world.</b>
